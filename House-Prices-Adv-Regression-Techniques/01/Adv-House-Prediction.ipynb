# FUNCTIONS

```python
def cramer_v(x,y):
    cm = pd.crosstab(x,y).values
    chi2 = stats.chi2_contingency( cm )[0]
    n = cm.sum()
    r, k = cm.shape
    chi2corr = max(0,chi2 - (k-1)*(r-1)/(n-1))
    kcorr = k - (k-1)**2/(n-1)
    
    rcorr = r - (r-1)**2/(n-1)
    
    return np.sqrt((chi2/n)/ (min(kcorr-1,rcorr-1)))

def mean_percentage_error(y, yhat):
    return np.mean((y - yhat) / y)


def mean_absolute_percentage_error(y, yhat):
    return np.mean(np.abs((y - yhat) / y))


def ml_error(model_name, y, yhat):
    mae = mean_absolute_error(y, yhat)
    mape = mean_absolute_percentage_error(y, yhat)
    rmse = np.sqrt(mean_squared_error(y, yhat))
    
    return pd.DataFrame({'Model Name' : model_name,
                        'MAE' : mae,
                        'MAPE' : mape,
                        'RMSE' : rmse}, index = [0])
```

# MACHINE LEARNING

## Average Model

```python
aux_mean = X_test.copy()

# sales column
aux_mean['PRICE'] = y_test.copy()

# AVG
aux_mean['AVG_Model'] = aux_mean['PRICE'].mean()

mean_result = ml_error('AVG Model', y_test, aux_mean['AVG_Model'])
mean_result
```

## Linear Regression Model

```python
# model
lr = LinearRegression().fit(X_train, y_train)

# prediction
yhat_lr = lr.predict(X_test)

#performance
lr_result = ml_error('Linear Regression', y_test, yhat_lr)
lr_result
```

### Linear Regression - Cross Validation

```python
lr_cross = cross_validate(lr, X_train, y_train, cv = 5)

lr_cv = lr_cross['test_score'].mean(), lr_cross['test_score'].std()
```

```python
cv_resultLR = pd.DataFrame(lr_cv, columns=['LR_CV'],
            index = ['Mean', 'STD'])

cv_resultLR
```

## Linear Regression Regularized Model - Lasso

```python
# model
lrr = Lasso(alpha = 5).fit(X_train, y_train)

# prediction
yhat_lrr = lrr.predict(X_test)

# performance
lrr_result = ml_error('Linear Regression - Lasso', y_test, yhat_lrr)
lrr_result
```

### Lasso - Cross Validation

```python
lrr_cross = cross_validate(lrr, X_train, y_train, cv = 5)

lrr_cv  = lrr_cross['test_score'].mean(), lrr_cross['test_score'].std()

cv_resultLRR = pd.DataFrame(lrr_cv, columns=['LRR_CV'],
            index = ['Mean', 'STD'])

cv_resultLRR
```

## Random Forest Regressor 

```python
# model
rf = RandomForestRegressor(n_estimators = 100, n_jobs = 1, random_state = 42).fit(X_train, y_train)

# prediction
yhat_rf = rf.predict(X_test)

# Performance
rf_result = ml_error('Random Forest Regressor', y_test, yhat_rf)
rf_result
```

### Random Forest Regressor - Cross Validation

```python
rf_cross = cross_validate(rf, X_train, y_train, cv = 5)

rf_cv = rf_cross['test_score'].mean(), rf_cross['test_score'].std()


cv_resultRF = pd.DataFrame(rf_cv, columns=['RF_CV'],
            index = ['Mean', 'STD'])

cv_resultRF
```

```python
xgb_cross = cross_validate(xgb_model, X_train, y_train, cv = 5)

xgb_cv = xgb_cross['test_score'].mean(), xgb_cross['test_score'].std()

cv_resultxgb = pd.DataFrame(xgb_cv, columns=['XGB_CV'],
            index = ['Mean', 'STD'])

cv_resultxgb
```
